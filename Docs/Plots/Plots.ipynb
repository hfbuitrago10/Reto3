```python
import matplotlib.pyplot as plt
```

# **Tiempo de ejecución**
Para la recolección de datos se promedió el tiempo de ejecución para cada requerimiento después de ser ejecutados 5 veces. Las siguientes gráficas muestran los resultados de las pruebas.

```python
x = [1, 2, 3, 4, 5]

y_1 = [221.120, 563.585,195.837,6468.104,4651.850]
y_2 = [2362.509,6033.646,1432.374,39935.196,30431.173]
y_3 = [6972.809,13125.987,2744.861,77830.861,61372.849]

x_labels = ["1", "2", "3", "4", "5"]

plt.plot(x, y_1, marker='x')
plt.plot(x, y_2, marker='^')
plt.plot(x, y_3, marker='.')

plt.xticks(x, x_labels)
plt.xlim([0.5,5.5])
plt.xlabel('Requerimiento')
plt.ylabel('log Tiempo [ms]')
plt.yscale('log')
plt.title('Tiempo de ejecución')
plt.legend(['small', '5-pct', '10-pct'], loc='upper left')

plt.show()
```
